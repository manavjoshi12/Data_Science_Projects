```python
import pandas as pd 
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns 
```

```python
netflix = pd.read_csv('netflix_titles.csv.zip')
```

```python
netflix
```

```python
netflix.isnull()
```

```python
netflix.info()
```

```python
netflix['date_added']= pd.to_datetime(netflix['date_added'] , errors='coerce')
```

```python
netflix.info()
```

```python
netflix.columns = netflix.columns.str.strip()
```

```python
netflix['year_added']= netflix['date_added'].dt.year
netflix['month_added']= netflix['date_added'].dt.month
```

```python
netflix
```

```python
netflix.isnull().sum()
```

```python
sns.countplot(data= netflix, x='type',palette='Set2')
plt.title('distribution of movies')
plt.xlabel('type')
plt.ylabel('count')
plt.show()
```

```python
netflix['year_added'].value_counts().sort_index().plot(kind='bar',figsize=(10,5),color='coral') 
plt.title('distribution of movies and shows as per the year')
plt.xlabel('Year')
plt.ylabel('No. of titles')
plt.xticks(rotation=45)
plt.show()
```

```python
netflix['country']= netflix['country'].fillna('Unknown')
top_countries = netflix['country'].value_counts().head(10)
top_countries.plot(kind='barh',color='teal')
plt.title('top 10 countries producing netflix content')
plt.xlabel('number of titles')
plt.gca().invert_yaxis()
plt.show()
```

```python
from collections import Counter
```

```python
all_genres= netflix['listed_in'].dropna().str.split(',')
genre_list = [genre for sublist in all_genres for genre in sublist ]
genre_counts= pd.Series(Counter(genre_list)).sort_values(ascending=False)
genre_counts.head(10).plot(kind= 'bar', color= 'teal')
plt.title('Top 10 genres in netflix')
plt.xlabel('Number of titles')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=netflix, y='rating', order=netflix['rating'].value_counts().index, palette='viridis')
plt.title("Distribution of Ratings")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

```python
heatmap_data = netflix.pivot_table(index='year_added', columns='month_added', values='show_id', aggfunc='count')
sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=0.5)
plt.title("Netflix Additions Heatmap (Year vs. Month)")
plt.xlabel("Month")
plt.ylabel("Year")
plt.show()
```

```python
netflix['director'] = netflix['director'].fillna('Unknown')
top_directors = netflix['director'].value_counts().head(10)

top_directors.plot(kind='bar', color='salmon')
plt.title("Top 10 Directors on Netflix")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```
